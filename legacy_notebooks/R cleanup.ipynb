```r
newdf = as.data.frame(train2$ProductId)
```

```r
head(newdf)
```

```r
tail(newdf)
```

```r
summ_c = gsub("\\<.*?\\>", " ", train2$Summary, perl=TRUE)
text_c = gsub("\\<.*?\\>", " ", train2$Text, perl=TRUE)
```

```r
head(newdf)
head(train2)
```
